```python
import pandas as pd
import numpy as np
```

```python
file_to_load = "Resources/purchase_data.csv"
purchase_data = pd.read_csv(file_to_load)
```

```python
total_players = len(purchase_data["SN"].value_counts())
player = pd.DataFrame({"Total Players" : [total_players]})
player
```

```python
unique_items = len(purchase_data["Item ID"].unique())
average_price = (purchase_data["Price"].mean())
total_purchases = (purchase_data["Purchase ID"].count())
total_revenue = (purchase_data["Price"].sum())

output = pd.DataFrame({"Number of Unique Items" : [unique_items],
                       "Average Purchase Price": [average_price], 
                       "Total Number of Purchases": [total_purchases],
                       "Total Revenue": [total_revenue]})
output
```

```python
gender_count = purchase_data["Gender"].value_counts()
gender_percentage = gender_count/total_players *100 

output_next = pd.DataFrame({"Total Count" : gender_count,
                      "Percentage of Players": gender_percentage})

output_next
```

```python
purchase_counts = purchase_data.groupby(["Gender"]).count()["Price"]
average_purchase_price = purchase_data.groupby(["Gender"]).mean()["Price"]
total_purchase_value = purchase_data.groupby(["Gender"]).sum()["Price"]
average_per_person = total_purchase_value / output_next["Total Count"]

output_third = pd.DataFrame({"Purchase Counts": purchase_counts,
               "Average Purchase Price": average_purchase_price,
                "Total Purchase Value": total_purchase_value,
                "Average Total Purchase per Person": average_per_person})
output_third
```

```python
bins = [0, 10, 15, 20, 25, 30, 35, 40, 45, 50, 200]
bin_labels = ["Younger than 10", "10-15", "16-20", "21-25", "26-30", "31-35", "36-40", "41-45", "46-50", "Older than 51"]
purchase_data["Age"] = pd.cut(purchase_data["Age"],bins, labels=bin_labels)
age_total_count = purchase_data["Age"].value_counts()
age_percentages = age_total_count / total_players * 100
output_fourth = pd.DataFrame({"Total Count": age_total_count, 
                              "Percentage of Players": age_percentages})
output_fourth.sort_index()
```

```python
# sum = total purchase value 
# mean = average purchase value 
# count = purchase count
age_purchase_counts = purchase_data.groupby(["Age"]).count()["Price"]
age_average_purchase_price = purchase_data.groupby(["Age"]).mean()["Price"]
age_total_purchase_value = purchase_data.groupby(["Age"]).sum()["Price"]
age_average_per_person = age_total_purchase_value/output_fourth["Total Count"]

output_fifth = pd.DataFrame({"Purchase Count" : age_purchase_counts,
                            "Average Purchase Price": age_average_purchase_price,
                            "Total Purchase Value": age_total_purchase_value,
                            "Avg total Purchase per Person": age_average_per_person})
output_fifth.sort_index()
```

```python
purchase_count = purchase_data.groupby(["SN"]).count()["Price"]
average_purchase_price = purchase_data.groupby(["SN"]).mean()["Price"]
total_purchase_value = purchase_data.groupby(["SN"]).sum()["Price"]

output_sixth = pd.DataFrame({"Purchase Count": purchase_count,
                             "Average Purchase Price": average_purchase_price,
                             "Total Purchase Value": total_purchase_value})
output_sixth.sort_values("Total Purchase Value", ascending=False).head(5)
```

```python
popular_items = purchase_data.loc[:,["Item ID", "Item Name", "Price"]]

total_purchase_value = popular_items.groupby(["Item ID", "Item Name"]).sum()["Price"]
item_price = popular_items.groupby(["Item ID", "Item Name"]).mean()["Price"]
purchase_count = popular_items.groupby(["Item ID", "Item Name"]).count()["Price"]

output_seventh = pd.DataFrame({"Purchase Count": purchase_count,
                              "Item Price": item_price,
                              "Total Purchase Value": total_purchase_value})
output_seventh.sort_values("Purchase Count", ascending=False).head(5)
```
